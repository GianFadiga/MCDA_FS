# MCDA: PROJECT - Final Algorithm

```python
import pandas as pd
import plotly.express as px
import numpy as np

class DataAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
        self.pesos = None
        self.tipos_dados = None
        self.proporcionalidade = None
        self.valores_bom = None
        self.valores_neutro = None
        self.calculo_df = None
        # Adicione o mapeamento de string aqui para ser acess칤vel
        self.string_columns_map = {}


    def load_and_prepare_data(self):
        """Carrega e prepara os dados de forma gen칠rica para strings"""
        self.df = pd.read_csv(self.file_path, encoding='UTF-8', sep=',', skipinitialspace=True)

        # Remove colunas totalmente vazias
        self.df = self.df.dropna(axis=1, how='all')

        # Extrai configura칞칫es ANTES de mapear colunas string
        self.pesos = self.df.iloc[0].dropna().astype(float)
        self.tipos_dados = self.df.iloc[1].dropna()
        self.proporcionalidade = self.df.iloc[2].dropna()

        # Identifica automaticamente colunas string e seus pts_string
        # 칄 importante fazer isso *depois* de ler os tipos_dados
        self._mapear_colunas_string()

        # Extrai BOM/NEUTRO
        # Usar .get para evitar KeyError se 'Modelo' n칚o for a primeira coluna
        bom_row = self.df[self.df.iloc[:, 0] == 'BOM']
        neutro_row = self.df[self.df.iloc[:, 0] == 'NEUTRO']

        if not bom_row.empty:
             self.valores_bom = bom_row.iloc[0].dropna()
        else:
             print("Aviso: Linha 'BOM' n칚o encontrada.")
             self.valores_bom = pd.Series(dtype='object')

        if not neutro_row.empty:
             self.valores_neutro = neutro_row.iloc[0].dropna()
        else:
             print("Aviso: Linha 'NEUTRO' n칚o encontrada.")
             self.valores_neutro = pd.Series(dtype='object')


        # Cria calculo_df pulando as 5 primeiras linhas (pesos, tipo, prop, BOM, NEUTRO)
        # Assume que 'Modelo' 칠 a primeira coluna para a l칩gica BOM/NEUTRO
        self.calculo_df = self.df.iloc[5:].reset_index(drop=True)
        self._convert_data_types()

    def _convert_data_types(self):
        """Converte os tipos de dados conforme especificado"""
        # Garante que valores_bom e valores_neutro sejam Series
        if self.valores_bom is None: self.valores_bom = pd.Series(dtype='object')
        if self.valores_neutro is None: self.valores_neutro = pd.Series(dtype='object')

        for col in self.tipos_dados.index:
            if col not in self.calculo_df.columns:
                print(f"Aviso: Coluna '{col}' definida nos tipos mas n칚o encontrada nos dados para c치lculo.")
                continue

            tipo = self.tipos_dados.get(col) # Usar .get para seguran칞a

            if tipo == 'number':
                # Tenta converter BOM/NEUTRO para float, usa NaN se falhar
                self.valores_bom[col] = pd.to_numeric(self.valores_bom.get(col), errors='coerce')
                self.valores_neutro[col] = pd.to_numeric(self.valores_neutro.get(col), errors='coerce')
                # Converte coluna de dados
                self.calculo_df[col] = pd.to_numeric(self.calculo_df[col], errors='coerce')

            elif tipo == 'boolean':
                 # --- CORRE칂츾O BOOLEAN ---
                 # Converte BOM/NEUTRO: 'TRUE' -> True, outros -> False
                 self.valores_bom[col] = str(self.valores_bom.get(col, '')).strip().upper() == 'TRUE'
                 self.valores_neutro[col] = str(self.valores_neutro.get(col, '')).strip().upper() == 'TRUE'
                 # Converte coluna de dados com mapeamento robusto
                 self.calculo_df[col] = self.calculo_df[col].astype(str).str.strip().str.upper().map({'TRUE': True, 'FALSE': False})
                 # Define um padr칚o para valores n칚o mapeados (ex: NaN ou outros textos)
                 self.calculo_df[col] = self.calculo_df[col].fillna(False).astype(bool) # Trata NaN e for칞a tipo bool

            elif tipo == 'string':
                # Apenas garante que a coluna de dados seja string
                self.calculo_df[col] = self.calculo_df[col].astype(str)

            # Lidar com colunas pts_string (n칚o precisam de convers칚o aqui)
            elif tipo == 'pts_string':
                 continue # Ignora pts_string na convers칚o principal

            else:
                 print(f"Aviso: Tipo de dado desconhecido '{tipo}' para coluna '{col}'.")


    def calcular_pontuacao_numerica(self, valor, bom, neutro, peso, proporcionalidade_tipo):
        """Calcula pontua칞칚o para dados num칠ricos e booleanos"""

        # --- CORRE칂츾O BOOLEAN LOGIC ---
        # Trata explicitamente valores booleanos primeiro
        # Usar isinstance com np.bool_ tamb칠m por causa do Pandas/Numpy
        if isinstance(valor, (bool, np.bool_)):
            # Caso especial: BOM e NEUTRO s칚o iguais (ambos True ou ambos False)
            if bom == neutro:
                return peso if valor == bom else 0 # Pontua se igual ao 칰nico alvo, sen칚o 0
            # Caso normal: BOM e NEUTRO s칚o diferentes
            else:
                if valor == bom:
                    return peso  # Acertou o BOM
                elif valor == neutro:
                    return 0     # Acertou o NEUTRO
                else:
                    # N칚o 칠 nem BOM nem NEUTRO (deve ser o oposto do BOM e do NEUTRO)
                    return -peso # Penalidade

        # Se n칚o for booleano, segue para a l칩gica num칠rica
        # Verifica se algum valor essencial 칠 NaN antes de prosseguir
        if pd.isna(valor) or pd.isna(bom) or pd.isna(neutro) or pd.isna(peso):
             return 0 # N칚o pontua se faltar informa칞칚o

        # Se BOM == NEUTRO para n칰meros, retorna 0 (ou peso se valor for igual?)
        # Mantendo a l칩gica original: retorna 0 se bom == neutro
        if bom == neutro:
            return 0

        # Tenta converter para float (j치 deveriam ser, mas por seguran칞a)
        try:
            valor = float(valor)
            bom = float(bom)
            neutro = float(neutro)
            peso = float(peso)
        except (ValueError, TypeError):
            # Se a convers칚o falhar aqui, algo est치 errado antes
            print(f"Aviso: Falha na convers칚o para float em calcular_pontuacao_numerica - Valor: {valor}, Bom: {bom}, Neutro: {neutro}")
            return 0

        # L칩gica de proporcionalidade (mantida como estava, mas agora com valores mais seguros)
        # Adicionando tratamento para divis칚o por zero ou neutro zero
        if proporcionalidade_tipo == 'proportional':
            if bom > neutro: # Caso normal crescente
                 if valor >= bom:
                     return peso
                 elif valor >= neutro:
                      # Evita divis칚o por zero se bom == neutro (j치 tratado acima)
                     return ((valor - neutro) / (bom - neutro)) * peso
                 else: # Abaixo do neutro
                     # Evita divis칚o por zero se neutro for 0 e valor negativo
                     denominador_penalidade = neutro if neutro != 0 else 1 # Usa 1 se neutro for 0
                     return -((abs(neutro - valor)) / abs(denominador_penalidade)) * peso # Penalidade proporcional  dist칙ncia do neutro
            elif bom < neutro: # Caso decrescente (ex: menor 칠 melhor at칠 BOM) - Incomum para 'proportional' mas poss칤vel
                 if valor <= bom:
                     return peso
                 elif valor <= neutro:
                     return ((valor - bom) / (neutro - bom)) * peso # Inverte a l칩gica
                 else: # Acima do neutro (pior)
                    denominador_penalidade = neutro if neutro != 0 else 1
                    return -((abs(valor - neutro)) / abs(denominador_penalidade)) * peso
            else: # bom == neutro
                 return 0

        elif proporcionalidade_tipo == 'i_proportional':
             if bom < neutro: # Caso normal decrescente (menor 칠 melhor)
                 if valor <= bom:
                     return peso
                 elif valor <= neutro:
                      # Evita divis칚o por zero se neutro == bom (j치 tratado acima)
                     return ((neutro - valor) / (neutro - bom)) * peso
                 else: # Acima do neutro (pior)
                     denominador_penalidade = neutro if neutro != 0 else 1
                     return -((abs(valor - neutro)) / abs(denominador_penalidade)) * peso # Penalidade proporcional
             elif bom > neutro: # Caso crescente (maior 칠 melhor at칠 BOM) - Incomum para 'i_proportional'
                 if valor >= bom:
                     return peso
                 elif valor >= neutro:
                     return ((bom - valor) / (bom - neutro)) * peso # Inverte a l칩gica
                 else: # Abaixo do neutro (pior)
                     denominador_penalidade = neutro if neutro != 0 else 1
                     return -((abs(neutro - valor)) / abs(denominador_penalidade)) * peso
             else: # bom == neutro
                 return 0

        return 0 # Caso tipo de proporcionalidade n칚o seja reconhecido


    def _mapear_colunas_string(self):
        """Mapeia automaticamente colunas string para suas colunas pts_string"""
        self.string_columns_map = {}
        if self.tipos_dados is None: # Garante que tipos_dados foi carregado
             print("Erro: tipos_dados n칚o carregado antes de mapear colunas string.")
             return

        string_cols = [col for col in self.tipos_dados.index
                      if self.tipos_dados[col] == 'string']

        df_cols = list(self.df.columns) # Lista das colunas originais

        for col in string_cols:
            try:
                # Encontra a posi칞칚o da coluna string
                col_idx = df_cols.index(col)

                # A coluna pts_string deve ser a pr칩xima coluna
                if col_idx + 1 < len(df_cols):
                    pts_potential_col = df_cols[col_idx + 1]
                    # Verifica se a pr칩xima coluna 칠 realmente a de pontos (opcional, mas bom)
                    # Poderia verificar se o tipo dela 칠 'pts_string' na linha 1, mas vamos confiar na posi칞칚o por enquanto
                    new_pts_col_name = f"{col},pts_string"
                    self.string_columns_map[col] = new_pts_col_name # Armazena o nome *novo*

                    # Renomeia a coluna no DataFrame original
                    self.df = self.df.rename(columns={pts_potential_col: new_pts_col_name})
                    print(f"Coluna string '{col}' mapeada para '{new_pts_col_name}' (original: '{pts_potential_col}')")

                else:
                    print(f"Aviso: Coluna string '{col}' n칚o tem uma coluna seguinte para ser pts_string.")
            except ValueError:
                 print(f"Aviso: Coluna string '{col}' definida nos tipos n칚o encontrada no CSV.")
            except Exception as e:
                 print(f"Erro ao mapear coluna string '{col}': {e}")


    def calcular_pontuacao_string(self, valor, coluna):
        """Calcula pontua칞칚o gen칠rica para qualquer string"""
        # Obt칠m o nome CORRETO da coluna de pontos, que foi renomeada
        pts_col = self.string_columns_map.get(coluna)

        if not pts_col:
            # Isso n칚o deveria acontecer se _mapear_colunas_string funcionou
            print(f"Aviso interno: Mapeamento para coluna string '{coluna}' n칚o encontrado.")
            return 0

        if pts_col not in self.df.columns:
             # Isso pode acontecer se o renomear falhou ou a coluna n칚o existia
             print(f"Aviso: Coluna de pontos '{pts_col}' esperada para '{coluna}' n칚o existe no DataFrame.")
             return 0

        try:
            # --- CORRE칂츾O STRING ---
            # 1. Criar o mapeamento a partir das linhas de dados do df original
            dados_mapa = self.df.iloc[5:].copy()

            # 2. Converter a coluna de pontos para num칠rico DEPOIS de selecionar os dados
            dados_mapa[pts_col] = pd.to_numeric(dados_mapa[pts_col], errors='coerce')

            # 3. Remover linhas onde a string original OU o ponto convertido s칚o NaN
            dados_mapa = dados_mapa.dropna(subset=[coluna, pts_col])

            # 4. Criar o dicion치rio de mapeamento (String -> Ponto Num칠rico)
            mapping = pd.Series(dados_mapa[pts_col].values, index=dados_mapa[coluna]).to_dict()

            # Se o mapping estiver vazio ap칩s a limpeza, retorna 0
            if not mapping:
                 print(f"Aviso: Mapeamento de pontua칞칚o para '{coluna}' est치 vazio ap칩s limpeza.")
                 return 0

            # Remove espa칞os e busca o valor no mapeamento
            valor_clean = str(valor).strip()
            # O valor buscado no dicion치rio J츼 칄 float (ou int)
            pontuacao_base = mapping.get(valor_clean, 0) # Pega o ponto base (j치 num칠rico)

            # Multiplica o ponto base pelo peso da coluna string
            peso_coluna = float(self.pesos.get(coluna, 0)) # Pega o peso com seguran칞a
            resultado = pontuacao_base * peso_coluna

            # Debugging extra (opcional):
            # print(f"  Debug String: Col: {coluna}, Valor: '{valor_clean}', Ponto Base: {pontuacao_base}, Peso: {peso_coluna}, Result: {resultado}")

            return resultado

        except Exception as e:
            # Captura erros mais espec칤ficos se poss칤vel
            print(f"Erro detalhado ao calcular pontua칞칚o para '{coluna}' (Valor: '{valor}'): {type(e).__name__} - {str(e)}")
            # Imprimir o estado do mapeamento pode ajudar
            # print(f"  Debug - Estado do mapping problem치tico: {mapping}")
            return 0

    def calcular_pontuacoes(self):
        """Calcula todas as pontua칞칫es com verifica칞칚o extra"""
        if self.calculo_df is None or self.tipos_dados is None or self.pesos is None:
             print("Erro: Dados n칚o carregados ou preparados corretamente. Execute load_and_prepare_data() primeiro.")
             return

        print("Colunas dispon칤veis para c치lculo:", list(self.calculo_df.columns))
        print("Pesos aplic치veis:", self.pesos.to_dict())
        print("Tipos de Dados:", self.tipos_dados.to_dict())


        for col in self.pesos.index:
            # Verifica se a coluna de peso existe nos dados de c치lculo
            if col not in self.calculo_df.columns:
                # Verifica se 칠 uma coluna string (cujo c치lculo usa a coluna original)
                if col not in self.string_columns_map:
                     print(f"  Aviso: Coluna de peso '{col}' n칚o encontrada no DataFrame de c치lculo e n칚o 칠 uma coluna string mapeada.")
                     continue
                # Se for string, o c치lculo usar치 a coluna original, ent칚o est치 OK n칚o estar em calculo_df diretamente,
                # mas precisamos garantir que o tipo 칠 'string'
                elif self.tipos_dados.get(col) != 'string':
                     print(f"  Aviso: Coluna de peso '{col}' n칚o encontrada no DataFrame de c치lculo, mas mapeada como string ({self.string_columns_map.get(col)}). Verifique a consist칡ncia.")
                     continue
                # Se for string mapeada, o processamento ocorrer치 no bloco 'string' abaixo

            print(f"\nProcessando coluna: {col}")

            # Pega informa칞칫es com seguran칞a usando .get()
            tipo = self.tipos_dados.get(col)
            proporcionalidade = self.proporcionalidade.get(col)
            peso = self.pesos.get(col, 0) # Usa 0 se peso n칚o for encontrado por algum motivo
            valor_bom = self.valores_bom.get(col)
            valor_neutro = self.valores_neutro.get(col)
            pont_col_name = f"{col}_pontuacao" # Nome da coluna de resultado

            try:
                if tipo in ['number', 'boolean']:
                    # Verifica se todas as informa칞칫es necess치rias est칚o presentes
                    if pd.isna(valor_bom) or pd.isna(valor_neutro):
                         print(f"  Aviso: Valores BOM ({valor_bom}) ou NEUTRO ({valor_neutro}) ausentes/inv치lidos para '{col}'. Pulando pontua칞칚o.")
                         self.calculo_df[pont_col_name] = 0
                         continue
                    if not proporcionalidade:
                         print(f"  Aviso: Proporcionalidade n칚o definida para '{col}'. Pulando pontua칞칚o.")
                         self.calculo_df[pont_col_name] = 0
                         continue

                    print(f"  Tipo: {tipo}, Proporcionalidade: {proporcionalidade}")
                    print(f"  Valores BOM: {valor_bom}, NEUTRO: {valor_neutro}, Peso: {peso}")

                    self.calculo_df[pont_col_name] = self.calculo_df.apply(
                        lambda row: self.calcular_pontuacao_numerica(
                            row[col],
                            valor_bom,
                            valor_neutro,
                            peso,
                            proporcionalidade
                        ) if col in row else 0, # Adiciona verifica칞칚o se col existe na linha
                        axis=1
                    )

                elif tipo == 'string':
                    # Verifica se a coluna original existe
                    if col not in self.calculo_df.columns:
                         print(f"  Erro Interno: Coluna string '{col}' n칚o encontrada em calculo_df para aplicar pontua칞칚o.")
                         self.calculo_df[pont_col_name] = 0
                         continue

                    pts_col_esperado = self.string_columns_map.get(col)
                    print(f"  Tipo: {tipo}. Coluna de pontos esperada: {pts_col_esperado}. Peso: {peso}")
                    # A valida칞칚o da exist칡ncia de pts_col 칠 feita dentro de calcular_pontuacao_string

                    self.calculo_df[pont_col_name] = self.calculo_df.apply(
                        lambda row: self.calcular_pontuacao_string(row[col], col),
                        axis=1
                    )

                # Imprime resultado parcial se a coluna foi criada
                if pont_col_name in self.calculo_df:
                    print(f"  Resultado parcial (Top 5):\n{self.calculo_df[[col, pont_col_name]].head().to_string(index=False)}")
                else:
                     print(f"  Aviso: Coluna de pontua칞칚o '{pont_col_name}' n칚o foi criada.")

            except KeyError as e:
                 print(f"  Erro de Chave (KeyError) ao processar coluna '{col}': {e}. Verifique se a coluna existe e os nomes est칚o corretos nas configura칞칫es e no CSV.")
                 self.calculo_df[pont_col_name] = 0
            except Exception as e:
                print(f"  Erro inesperado ao processar coluna '{col}': {type(e).__name__} - {str(e)}")
                import traceback
                # traceback.print_exc() # Descomente para obter traceback completo se necess치rio
                self.calculo_df[pont_col_name] = 0

        # Calcula pontua칞칚o total
        colunas_pontuacao = [c for c in self.calculo_df.columns if c.endswith('_pontuacao')]
        if not colunas_pontuacao:
             print("\nNenhuma coluna de pontua칞칚o foi gerada. Pontua칞칚o total n칚o calculada.")
             self.calculo_df['Pontuacao_Total'] = 0
        else:
             print(f"\nColunas usadas para Pontua칞칚o Total: {colunas_pontuacao}")
             self.calculo_df['Pontuacao_Total'] = self.calculo_df[colunas_pontuacao].sum(axis=1)

        print("\nResumo final das pontua칞칫es:")
        # Seleciona colunas de forma segura
        cols_to_show = [self.calculo_df.columns[0]] + colunas_pontuacao + ['Pontuacao_Total']
        valid_cols_to_show = [c for c in cols_to_show if c in self.calculo_df.columns]
        print(self.calculo_df[valid_cols_to_show].head().to_string(index=False))


  # Mostra as pontua칞칫es em forma de c칩digo (debug apenas)
    def mostrar_pontuacoes(self, apenas_pontuacoes=True):
      """Exibe o DataFrame com as pontua칞칫es calculadas

      Args:
          apenas_pontuacoes (bool): Se True, mostra apenas colunas de pontua칞칚o
                                    Se False, mostra todas as colunas
      """
      if self.calculo_df is None:
           print("DataFrame de c치lculo n칚o dispon칤vel.")
           return

      try:
           # Tenta importar display dinamicamente se estiver em um notebook
           from IPython.display import display
      except ImportError:
           # Se n칚o estiver em um notebook, usa print
           display = print

      if apenas_pontuacoes:
          # Filtra apenas colunas que terminam com '_pontuacao' ou s칚o 'Modelo' ou 'Pontuacao_Total'
          # Assume que a primeira coluna 칠 o identificador
          modelo_col = self.calculo_df.columns[0]
          cols_pontuacao = [c for c in self.calculo_df.columns if c.endswith('_pontuacao')]
          cols_total = ['Pontuacao_Total'] if 'Pontuacao_Total' in self.calculo_df.columns else []
          cols = [modelo_col] + cols_pontuacao + cols_total
          valid_cols = [c for c in cols if c in self.calculo_df.columns]
          display(self.calculo_df[valid_cols])
      else:
          display(self.calculo_df)

    def gerar_graficos(self):
        """Gera os gr치ficos de an치lise"""
        print("\n--- Gerando Gr치ficos ---")
        if self.calculo_df is None:
            print("DataFrame de c치lculo n칚o dispon칤vel para gerar gr치ficos.")
            return
        try:
            self._gerar_grafico_barras()  # Gr치fico de pontua칞칚o total
            self._gerar_grafico_numericos()  # Gr치ficos para cada coluna num칠rica
            self._gerar_grafico_strings()
            self._gerar_grafico_booleanos()
            print("--- Gr치ficos Gerados ---")
        except Exception as e:
            print(f"Erro ao gerar gr치ficos: {e}")
            import traceback
            traceback.print_exc()

    def _gerar_grafico_numericos(self):
        """Gr치fico de barras para colunas num칠ricas com verifica칞칚o de proporcionalidade"""
        if self.tipos_dados is None:
            print("Tipos de dados n칚o carregados, pulando gr치fico de num칠ricos.")
            return

        # Identifica colunas num칠ricas
        numeric_cols = [col for col in self.tipos_dados.index
                      if self.tipos_dados[col] == 'number' and col in self.calculo_df.columns]

        if not numeric_cols:
            print("Nenhuma coluna num칠rica encontrada para gerar gr치ficos.")
            return

        modelo_col = self.calculo_df.columns[0]  # Assume primeira coluna como identificador

        for col_num in numeric_cols:
            print(f"Gerando gr치fico para coluna num칠rica: '{col_num}'")

            # Prepara dados
            df_num = self.calculo_df[[modelo_col, col_num]].copy()
            df_num = df_num[~df_num[modelo_col].isin(['BOM', 'NEUTRO'])]
            df_num.dropna(subset=[col_num], inplace=True)

            if df_num.empty:
                print(f"Nenhum dado v치lido para plotar no gr치fico de '{col_num}'.")
                continue

            # Obt칠m valores de refer칡ncia e tipo de proporcionalidade
            valor_bom = self.valores_bom.get(col_num)
            valor_neutro = self.valores_neutro.get(col_num)
            proporcionalidade = self.proporcionalidade.get(col_num, '')

            # Verifica se 칠 inversamente proporcional
            inversamente_proporcional = proporcionalidade.lower() == 'i_proportional'

            # Adiciona informa칞칚o ao t칤tulo
            titulo = f"{col_num} {'(Inversamente Proporcional)' if inversamente_proporcional else ''}"

            # Classifica como positivo/negativo considerando a proporcionalidade
            df_num['Color_Category'] = df_num.apply(
                lambda x: self._classificar_valor(x[col_num], valor_bom, valor_neutro, inversamente_proporcional),
                axis=1
            )

            # Define o mapa de cores
            color_map = {
                'Positiva': 'lightgreen',
                'Negativa': 'palevioletred',
                'Neutra': 'lightgray'
            }

            # Ordena por valor (maior no topo)
            df_num_sorted = df_num.sort_values(col_num, ascending=not inversamente_proporcional)

            # Cria gr치fico
            fig = px.bar(df_num_sorted,
                        x=col_num,
                        y=modelo_col,
                        orientation='h',
                        color='Color_Category',
                        color_discrete_map=color_map,
                        hover_name=modelo_col,
                        hover_data={col_num: ':.2f'},
                        title=titulo)

            # Configura칞칫es de layout
            fig.update_layout(
                yaxis={'categoryorder': 'total ascending'},
                plot_bgcolor='white',
                paper_bgcolor='white',
                legend_title_text='Resultado',
                xaxis_title=col_num,
                yaxis_title='Modelo'
            )

            # Adiciona linhas de refer칡ncia e anota칞칫es
            if pd.notna(valor_neutro):
                self._adicionar_referencia(fig, float(valor_neutro), len(df_num_sorted),
                                        "M칤nimo Aceit치vel<br>(NEUTRO)", inversamente_proporcional)

            if pd.notna(valor_bom):
                self._adicionar_referencia(fig, float(valor_bom), len(df_num_sorted),
                                        "Desej치vel (BOM)", inversamente_proporcional)

            print(f"  Gr치fico de {col_num} pronto para exibi칞칚o.")
            fig.show()

    def _classificar_valor(self, valor, valor_bom, valor_neutro, inversamente_proporcional):
        """Classifica o valor como Positiva, Negativa ou Neutra"""
        if pd.isna(valor_neutro):
            return 'Neutra'

        valor = float(valor)
        valor_neutro = float(valor_neutro)

        if inversamente_proporcional:
            return 'Positiva' if valor <= valor_neutro else 'Negativa'
        else:
            return 'Positiva' if valor >= valor_neutro else 'Negativa'

    def _adicionar_referencia(self, fig, valor, y_pos, texto, inversamente_proporcional):
        """Adiciona linha de refer칡ncia e anota칞칚o ao gr치fico"""
        fig.add_vline(x=valor, line_dash='dash', line_color='black')

        fig.add_annotation(
            x=valor,
            y=y_pos - 1,
            xref="x",
            yref="y",
            text=texto,
            showarrow=True,
            font=dict(size=16, color="#000000"),
            align="center",
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            arrowcolor="#636363",
            ax=0,
            ay=-45,
            bordercolor="#c7c7c7",
            borderwidth=2,
            borderpad=4,
            bgcolor='white',
            opacity=1
        )

    def _gerar_grafico_barras(self):
        """Gr치fico de barras para pontua칞칚o total com linha ideal fixa e indica칞칚o do melhor produto."""
        print("Gerando gr치fico de barras...")
        # Garante que colunas e pesos existem
        if 'Pontuacao_Total' not in self.calculo_df.columns:
            print("Coluna 'Pontuacao_Total' n칚o encontrada. Pulando gr치fico de barras.")
            return
        # Verifica se os pesos foram carregados para calcular o m치ximo te칩rico
        if self.pesos is None or self.pesos.empty:
             print("Pesos n칚o definidos. N칚o 칠 poss칤vel calcular pontua칞칚o m치xima te칩rica. Usando m치ximo atingido.")
             # Como fallback, podemos usar o m치ximo atingido ou um valor fixo como 1.0
             # Vamos definir como 1.0 se os pesos n칚o estiverem dispon칤veis.
             max_score_teorico = 1.0
        else:
            # Calcula a pontua칞칚o m치xima te칩rica (soma dos pesos)
            max_score_teorico = self.pesos.sum()
            print(f"  Pontua칞칚o m치xima te칩rica calculada (soma dos pesos): {max_score_teorico:.2f}")

        modelo_col = self.calculo_df.columns[0] # Assume primeira coluna como identificador

        df_analise = self.calculo_df[[modelo_col, 'Pontuacao_Total']].copy()
        # Remove linhas BOM/NEUTRO
        df_analise = df_analise[~df_analise[modelo_col].isin(['BOM', 'NEUTRO'])]

        # Remove linhas com pontua칞칚o NaN antes de prosseguir e ordenar
        df_analise.dropna(subset=['Pontuacao_Total'], inplace=True)

        if df_analise.empty:
             print("Nenhum dado v치lido para plotar no gr치fico de barras ap칩s remover BOM/NEUTRO/NaN.")
             return

        df_analise['Color_Category'] = df_analise.apply(
            lambda x: ('Positiva' if x['Pontuacao_Total'] >= 0 else 'Negativa'),
            axis=1
        )

        color_map = {
            'Positiva': 'lightgreen',
            'Negativa': 'palevioletred'
        }

        # Ordena por pontua칞칚o DESCENDENTE para plotagem e identifica칞칚o do melhor
        df_analise_sorted = df_analise.sort_values('Pontuacao_Total', ascending=False)

        # Identifica o melhor produto (primeira linha ap칩s ordenar)
        # Adiciona verifica칞칚o caso df_analise_sorted fique vazio ap칩s dropna
        if df_analise_sorted.empty:
             print("Nenhum produto restante ap칩s limpeza para identificar o melhor.")
             top_product_name = "N/A"
             top_product_score = pd.NA
        else:
             top_product = df_analise_sorted.iloc[0]
             top_product_name = top_product[modelo_col]
             top_product_score = top_product['Pontuacao_Total']
             print(f"  Melhor produto identificado: {top_product_name} (Score: {top_product_score:.2f})")


        fig = px.bar(df_analise_sorted,
                    x='Pontuacao_Total',
                    y=modelo_col, # Modelo no eixo Y
                    orientation='h',
                    color='Color_Category',
                    color_discrete_map=color_map,
                    hover_name=modelo_col,
                    hover_data={'Pontuacao_Total':':.2f'}, # Formata hover
                    title='Pontua칞칚o Total dos Modelos')

        # Atualiza layout - Ordena eixo Y pela pontua칞칚o (melhor em cima)
        fig.update_layout(
            yaxis={'categoryorder': 'total ascending'},
            plot_bgcolor='white',
            paper_bgcolor='white',
            legend_title_text='Resultado',
            xaxis_title='Pontua칞칚o Total',
            yaxis_title='Modelo'
        )

        # Linha de refer칡ncia no neutro
        fig.add_vline(x=0, line_dash='dash', line_color='black')

        # Linha de refer칡ncia no desej치vel
        fig.add_vline(x=max_score_teorico, line_dash='dash', line_color='black')

        # Linha Desej치vel (BOM)
        fig.add_annotation(
            x=max_score_teorico,
            y=len(df_analise_sorted) - 1,
            xref="x",
            yref="y",
            text="Desej치vel (BOM)<br>(NEUTRO)",
            showarrow=True,
            font=dict(size=16, color="#000000"),
            align="center",
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            arrowcolor="#636363",
            ax=0,
            ay=-45,
            bordercolor="#c7c7c7",
            borderwidth=2,
            borderpad=4,
            bgcolor='white',
            opacity=1
        )

        # Linha M칤nimo Aceit치vel (NEUTRO)
        fig.add_annotation(
            x=0,
            y=len(df_analise_sorted) - 1,
            xref="x",
            yref="y",
            text="M칤nimo Aceit치vel<br>(NEUTRO)",
            showarrow=True,
            font=dict(size=16, color="#000000"),
            align="center",
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            arrowcolor="#636363",
            ax=0,
            ay=-45,
            bordercolor="#c7c7c7",
            borderwidth=2,
            borderpad=4,
            bgcolor='white',
            opacity=1
        )

        # --- Anota칞칚o para Melhor Produto ---
        # S칩 adiciona a anota칞칚o se identificamos um melhor produto v치lido
        if top_product_name != "N/A" and pd.notna(top_product_score):
            fig.add_annotation(
                    x=top_product_score,    # Posi칞칚o X na pontua칞칚o do melhor
                    y=top_product_name,     # Posi칞칚o Y no nome do melhor (categoria no eixo y)
                    text="游끥 Melhor Produto", # O texto da anota칞칚o
                    showarrow=False,        # SEM seta, como voc칡 definiu
                    font=dict(
                        color="#00008B",    # Cor da fonte (Azul Escuro)
                        size=12
                    ),
                    # align="center",         # Remova ou comente esta linha

                    # --- Adi칞칫es para Fundo e Borda ---
                    bgcolor="white",        # Cor do fundo = branco
                    bordercolor="black",    # Cor da borda = preta
                    borderwidth=1,        # Largura da borda = 1 pixel
                    borderpad=4,          # Espa칞amento interno = 4 pixels
                    # --- Ajuste de Posi칞칚o (sem seta) ---
                    # Usamos xshift/yshift para mover o *texto* em rela칞칚o ao ponto (x,y)
                    # Ajuste esses valores para posicionar onde desejar (ex: acima e  direita da ponta da barra)
                    xshift=-15,            # Experimente com valores negativos (ajuste conforme necess치rio)
                    yshift=25
            )

        # As anota칞칫es antigas que voc칡 colou n칚o s칚o necess치rias pois usamos annotation_text nas vlines
        # e adicionamos uma anota칞칚o espec칤fica para o melhor produto.

        print("  Gr치fico de barras pronto para exibi칞칚o.")
        fig.show()


    def _gerar_grafico_strings(self):
        """Gr치fico melhorado para vari치veis de string - agora para todas as colunas string"""
        if self.tipos_dados is None:
            print("Tipos de dados n칚o carregados, pulando gr치fico de strings.")
            return

        string_cols = [col for col in self.tipos_dados.index
                      if self.tipos_dados.get(col) == 'string' and col in self.calculo_df.columns]

        if not string_cols:
            return

        modelo_col = self.calculo_df.columns[0] # Primeira coluna como identificador

        for col_string in string_cols:
            print(f"Gerando gr치fico para string: '{col_string}'")
            pts_col_name = self.string_columns_map.get(col_string)

            if not pts_col_name or pts_col_name not in self.df.columns:
                print(f"Aviso: Coluna de pontos '{pts_col_name}' para '{col_string}' n칚o encontrada. Pulando gr치fico.")
                continue

            try:
                # Cria mapeamento de pontos base (String -> Ponto num칠rico)
                dados_mapa = self.df.iloc[5:].copy()
                dados_mapa[pts_col_name] = pd.to_numeric(dados_mapa[pts_col_name], errors='coerce')
                dados_mapa = dados_mapa.dropna(subset=[col_string, pts_col_name])
                valores_string_base = pd.Series(dados_mapa[pts_col_name].values, index=dados_mapa[col_string]).to_dict()

                if not valores_string_base:
                    print(f"Aviso: Mapeamento de pontos base para '{col_string}' vazio. Pulando gr치fico.")
                    continue

            except Exception as e:
                print(f"Erro ao criar mapeamento para gr치fico de string '{col_string}': {e}.")
                continue

            df_string = self.calculo_df[[modelo_col, col_string]].copy()
            df_string = df_string[~df_string[modelo_col].isin(['BOM', 'NEUTRO'])]

            if df_string.empty:
                print(f"Nenhum dado para plotar no gr치fico de string para '{col_string}'.")
                continue

            # Mapeia para obter o PONTO BASE para o gr치fico (n칚o a pontua칞칚o final ponderada)
            df_string['ValorBase'] = df_string[col_string].map(valores_string_base).fillna(0)
            df_string = df_string.sort_values('ValorBase', ascending=False)
            df_string['Posicao'] = range(len(df_string)) # Posi칞칚o X baseada na ordena칞칚o

            fig = px.scatter(df_string,
                            x='Posicao',
                            y='ValorBase', # Usa o valor base no eixo Y
                            color=col_string, # Cor baseada na categoria string
                            hover_name=modelo_col, # O que aparece ao passar o mouse
                            title=f'Compara칞칚o de {col_string.capitalize()} entre Modelos (Pontua칞칚o Base)',
                            size_max=15, # Tamanho dos pontos
                            hover_data={'Posicao': False, 'ValorBase': True, col_string: True},
                            labels={'ValorBase': 'Pontua칞칚o Base (do CSV)', col_string: col_string.capitalize()})

            # Adiciona linhas BOM/NEUTRO com base nos valores BASE do mapeamento
            valor_bom_str = self.valores_bom.get(col_string)
            valor_neutro_str = self.valores_neutro.get(col_string)
            valor_base_bom = valores_string_base.get(valor_bom_str, None) # Busca ponto base do BOM
            valor_base_neutro = valores_string_base.get(valor_neutro_str, None) # Busca ponto base do NEUTRO

            if valor_base_bom is not None:
                fig.add_hline(y=valor_base_bom, line_dash='dash',
                          line_color='green', line_width=2,
                          annotation_text=f"BOM: '{valor_bom_str}' ({valor_base_bom:.2f})",
                          annotation_position="top right",
                          annotation_font=dict(color='green', size=12))

            if valor_base_neutro is not None:
                fig.add_hline(y=valor_base_neutro, line_dash='dash',
                          line_color='orange', line_width=2,
                          annotation_text=f"NEUTRO: '{valor_neutro_str}' ({valor_base_neutro:.2f})",
                          annotation_position="bottom right",
                          annotation_font=dict(color='orange', size=12))

            fig.update_layout(
                xaxis=dict(title='Modelos Ordenados por Pontua칞칚o Base', showticklabels=False, showgrid=False, zeroline=False),
                yaxis=dict(title='Pontua칞칚o Base (Definida no CSV)', showgrid=True, gridcolor='lightgray', zeroline=True, zerolinecolor='lightgray'),
                plot_bgcolor='white', paper_bgcolor='white',
                margin=dict(l=40, r=40, t=60, b=40),
                legend=dict(title_text=col_string.capitalize(), orientation='h', yanchor='bottom', y=-0.3, xanchor='center', x=0.5),
                hoverlabel=dict(bgcolor='white', font_size=12, font_family='Arial')
            )

            # CORRE칂츾O: Removido o par칙metro 'selector' que causava o erro
            fig.update_traces(marker=dict(size=12, line=dict(width=1, color='DarkSlateGrey'), opacity=0.8))

            fig.show()

    def _gerar_grafico_booleanos(self):
        """Gr치fico para todas as vari치veis booleanas"""
        # Identifica colunas booleanas
        bool_cols = [col for col in self.tipos_dados.index
                    if self.tipos_dados[col] == 'boolean' and col in self.calculo_df.columns]

        if not bool_cols:
            return

        modelo_col = self.calculo_df.columns[0]

        for col_bool in bool_cols:
            df_bool = self.calculo_df[~self.calculo_df[modelo_col].isin(['BOM', 'NEUTRO'])][[modelo_col, col_bool]].copy()
            df_bool['Valor'] = df_bool[col_bool].map({True: 1, False: 0})
            df_bool['Tamanho'] = 20

            fig = px.scatter(df_bool,
                    x=modelo_col,
                    y='Valor',
                    color=col_bool,
                    color_discrete_map={True: '#4CAF50', False: '#F44336'},
                    size='Tamanho',
                    title=f'{col_bool} por Modelo',
                    hover_name=modelo_col,
                    hover_data={'Valor': False, 'Tamanho': False},
                    category_orders={col_bool: [True, False]})

            # Atualiza a legenda para mostrar SIM/N츾O
            fig.for_each_trace(lambda t: t.update(name='SIM' if t.name == 'True' else 'N츾O'))

            fig.update_yaxes(
                range=[-0.5, 1.5],
                tickvals=[0, 1],
                ticktext=['N칚o', 'Sim'],
                showgrid=False
            )

            # Adiciona linhas de refer칡ncia SIM/N츾O
            fig.add_hline(y=1, line_dash='dash', line_color='green')
            fig.add_hline(y=0, line_dash='dash', line_color='orange')

            # Altera apar칡ncia do gr치fico
            fig.update_layout(
                xaxis_title=None,
                yaxis_title=None,
                plot_bgcolor='white',
                paper_bgcolor='white'
            )

            # Linhas de refer칡ncia
            valor_bom = 1 if self.valores_bom[col_bool] else 0
            valor_neutro = 1 if self.valores_neutro[col_bool] else 0

            fig.show()

    def recomendar_produtos(self):
        """Recomenda produtos com base na pontua칞칚o incluindo todos os crit칠rios"""
        # Assume que a primeira coluna 칠 o identificador 'Modelo'
        modelo_col = self.df.columns[0]
        if 'Pontuacao_Total' not in self.calculo_df.columns:
            print("Pontua칞칚o total n칚o calculada. N칚o 칠 poss칤vel recomendar.")
            return

        produtos = self.calculo_df[~self.calculo_df[modelo_col].isin(['BOM', 'NEUTRO'])]

        if produtos.empty:
            print("Nenhum produto encontrado para recomenda칞칚o (ap칩s excluir BOM/NEUTRO).")
            return

        # Verifica se h치 pontua칞칫es v치lidas
        if produtos['Pontuacao_Total'].isna().all():
            print("Todas as pontua칞칫es totais s칚o inv치lidas. N칚o 칠 poss칤vel recomendar.")
            return

        pontuacao_maxima = produtos['Pontuacao_Total'].max()
        # Lida com o caso de todas as pontua칞칫es serem NaN ou n칚o haver m치ximo
        if pd.isna(pontuacao_maxima):
             print("N칚o foi poss칤vel determinar a pontua칞칚o m치xima v치lida.")
             return

        recomendados = produtos[produtos['Pontuacao_Total'] == pontuacao_maxima]

        print("\n--- Produtos Recomendados ---")
        if recomendados.empty:
             print("Nenhum produto atingiu a pontua칞칚o m치xima (pode indicar erro ou dados uniformes).")
             return

        for _, produto in recomendados.iterrows():
            print(f"\nProduto: {produto[modelo_col]} com pontua칞칚o total de {produto['Pontuacao_Total']:.2f}")
            print("Detalhes da Pontua칞칚o por Crit칠rio (Ordenado por Peso):")

            # Ordena crit칠rios por peso (do maior para o menor)
            # Garante que o peso existe antes de tentar ordenar
            criterios_ordenados = sorted(
                [(col, self.pesos[col]) for col in self.pesos.index if col in self.pesos],
                key=lambda item: item[1], reverse=True
            )

            for col, peso in criterios_ordenados:
                pont_col = f"{col}_pontuacao"
                # Verifica se a coluna de pontua칞칚o existe no produto e tem valor v치lido
                if pont_col not in produto or pd.isna(produto[pont_col]):
                    # print(f"  - {col}: Pontua칞칚o n칚o dispon칤vel ou inv치lida.")
                    continue # Pula se n칚o houver pontua칞칚o para este crit칠rio

                pontuacao = produto[pont_col]
                valor_atual = produto.get(col, "N/A") # Pega o valor original com seguran칞a
                tipo = self.tipos_dados.get(col, "desconhecido")

                # Formata a sa칤da para melhor leitura
                justificativa = f"(Peso: {peso:.2f}, Pontos: {pontuacao:.2f})"

                if tipo == 'string':
                    valor_atual_str = f"'{valor_atual}'" if valor_atual != "N/A" else valor_atual
                    if pontuacao > 0:
                        print(f"  - {col}: {valor_atual_str} [Vantagem] {justificativa}")
                    elif pontuacao < 0:
                         print(f"  - {col}: {valor_atual_str} [Desvantagem] {justificativa}")
                    else:
                         print(f"  - {col}: {valor_atual_str} [Neutro] {justificativa}")

                elif tipo == 'boolean':
                     valor_atual_str = "Sim" if valor_atual == True else ("N칚o" if valor_atual == False else "N/A")
                     if pontuacao > 0:
                          print(f"  - {col}: {valor_atual_str} [Vantagem] {justificativa}")
                     elif pontuacao < 0:
                          print(f"  - {col}: {valor_atual_str} [Desvantagem] {justificativa}")
                     else:
                          print(f"  - {col}: {valor_atual_str} [Neutro] {justificativa}")

                elif tipo == 'number':
                     valor_atual_str = f"{valor_atual:.1f}" if isinstance(valor_atual, (int, float)) else str(valor_atual)
                     valor_bom = self.valores_bom.get(col)
                     valor_neutro = self.valores_neutro.get(col)

                     desc = f"{valor_atual_str}"
                     if pd.notna(valor_bom) and pd.notna(valor_neutro):
                          desc += f" (Bom: {valor_bom:.1f}, Neutro: {valor_neutro:.1f})"

                     if pontuacao >= peso * 0.99: # Praticamente atingiu o m치ximo
                          print(f"  - {col}: {desc} [칍timo] {justificativa}")
                     elif pontuacao > 0:
                          print(f"  - {col}: {desc} [Bom] {justificativa}")
                     elif pontuacao == 0:
                           print(f"  - {col}: {desc} [Neutro] {justificativa}")
                     else: # pontuacao < 0
                          print(f"  - {col}: {desc} [Abaixo do Neutro] {justificativa}")
                else:
                     # Caso tipo desconhecido
                     print(f"  - {col}: {valor_atual} [Tipo Desconhecido] {justificativa}")


# --- Exemplo de Uso - Notebook ---
# Certifique-se que o arquivo 'base_eletronicos_2.csv' est치 no mesmo diret칩rio
# ou forne칞a o caminho completo.
try:
    print("--- An치lise para base de eletr칪nicos ---")
    analyzer_eletro = DataAnalyzer('base_eletronicos_2.csv')
    analyzer_eletro.load_and_prepare_data()
    analyzer_eletro.calcular_pontuacoes()
    analyzer_eletro.gerar_graficos()
    analyzer_eletro.recomendar_produtos()
except FileNotFoundError:
    print("\nErro: Arquivo 'base_eletronicos_2.csv' n칚o encontrado.")
except Exception as e:
    print(f"\nOcorreu um erro inesperado durante a an치lise: {e}")
    import traceback
    traceback.print_exc()

# --- Exemplo de Uso - Carro ---
# Certifique-se que o arquivo 'base_eletronicos_2.csv' est치 no mesmo diret칩rio
# ou forne칞a o caminho completo.

# try:
#     print("--- An치lise para base de carros ---")
#     analyzer_carro = DataAnalyzer('base_carro.csv')
#     analyzer_carro.load_and_prepare_data()
#     analyzer_carro.calcular_pontuacoes()
#     analyzer_carro.gerar_graficos()
#     analyzer_carro.recomendar_produtos()
# except FileNotFoundError:
#     print("\nErro: Arquivo 'base_carro.csv' n칚o encontrado.")
# except Exception as e:
#     print(f"\nOcorreu um erro inesperado durante a an치lise: {e}")
#     import traceback
#     traceback.print_exc()
```

```python
"""
M칩dulo de an치lise de dados para compara칞칚o de produtos.

Este m칩dulo permite carregar, analisar e visualizar dados de compara칞칚o de produtos
com base em crit칠rios pr칠-definidos, gerando pontua칞칫es e recomenda칞칫es.
"""

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from typing import Dict, Optional, List, Union, Any
from fpdf import FPDF
import tempfile
import os
from PIL import Image
import base64

class DataAnalyzer:
    """Classe principal para an치lise e compara칞칚o de dados de produtos."""

    def __init__(self, file_path: str) -> None:
        self.file_path = file_path
        self.df = None
        self.weights = None
        self.data_types = None
        self.proportionality = None
        self.good_values = None
        self.neutral_values = None
        self.calculation_df = None
        self.string_columns_map = {}
        self.model_column = None

    def load_and_prepare_data(self) -> None:
        """Carrega e prepara os dados para an치lise."""
        self._load_raw_data()
        self._clean_empty_columns()
        self._extract_configurations()
        self._map_string_columns()
        self._extract_reference_values()
        self._prepare_calculation_data()
        self._convert_data_types()

    def _load_raw_data(self) -> None:
        """Carrega os dados brutos do arquivo CSV."""
        try:
            self.df = pd.read_csv(
                self.file_path,
                encoding='UTF-8',
                sep=',',
                skipinitialspace=True
            )
            self.model_column = self.df.columns[0]
        except Exception as e:
            raise ValueError(f"Erro ao carregar arquivo: {str(e)}")

    def _clean_empty_columns(self) -> None:
        """Remove colunas totalmente vazias do DataFrame."""
        self.df = self.df.dropna(axis=1, how='all')

    def _extract_configurations(self) -> None:
        """Extrai configura칞칫es de pesos, tipos de dados e proporcionalidade."""
        if len(self.df) < 3:
            raise ValueError("Arquivo CSV n칚o cont칠m linhas suficientes para configura칞칫es")

        self.weights = self.df.iloc[0].dropna().astype(float)
        self.data_types = self.df.iloc[1].dropna()
        self.proportionality = self.df.iloc[2].dropna()

    def _map_string_columns(self) -> None:
        """Mapeia colunas de string para suas colunas de pontos correspondentes."""
        self.string_columns_map = {}

        if self.data_types is None:
            raise ValueError("Tipos de dados n칚o foram carregados corretamente")

        string_cols = [
            col for col in self.data_types.index
            if self.data_types[col] == 'string'
        ]

        for col in string_cols:
            try:
                col_idx = list(self.df.columns).index(col)
                if col_idx + 1 < len(self.df.columns):
                    pts_col = self.df.columns[col_idx + 1]
                    new_pts_col_name = f"{col}_points"
                    self.string_columns_map[col] = new_pts_col_name
                    self.df = self.df.rename(columns={pts_col: new_pts_col_name})
            except (ValueError, IndexError):
                print(f"Aviso: N칚o foi poss칤vel mapear coluna string '{col}'")

    def _extract_reference_values(self) -> None:
        """Extrai valores de refer칡ncia BOM e NEUTRO."""
        good_row = self.df[self.df[self.model_column] == 'BOM']
        neutral_row = self.df[self.df[self.model_column] == 'NEUTRO']

        self.good_values = good_row.iloc[0].dropna() if not good_row.empty else pd.Series(dtype='object')
        self.neutral_values = neutral_row.iloc[0].dropna() if not neutral_row.empty else pd.Series(dtype='object')

        if good_row.empty or neutral_row.empty:
            print("Aviso: Linhas de refer칡ncia 'BOM' ou 'NEUTRO' n칚o encontradas")

    def _prepare_calculation_data(self) -> None:
        """Prepara o DataFrame para c치lculos, removendo linhas de configura칞칚o."""
        self.calculation_df = self.df.iloc[5:].reset_index(drop=True)

    def _convert_data_types(self) -> None:
        """Converte os tipos de dados conforme especificado."""
        for col in self.data_types.index:
            if col not in self.calculation_df.columns:
                continue

            dtype = self.data_types[col]

            try:
                if dtype == 'number':
                    self._convert_numeric_column(col)
                elif dtype == 'boolean':
                    self._convert_boolean_column(col)
                elif dtype == 'string':
                    self.calculation_df[col] = self.calculation_df[col].astype(str)
                elif dtype == 'pts_string':
                    continue
                else:
                    print(f"Aviso: Tipo de dado desconhecido '{dtype}' para coluna '{col}'")
            except Exception as e:
                print(f"Erro ao converter coluna '{col}': {str(e)}")

    def _convert_numeric_column(self, col: str) -> None:
        """Converte coluna num칠rica e valores de refer칡ncia."""
        self.good_values[col] = pd.to_numeric(self.good_values.get(col), errors='coerce')
        self.neutral_values[col] = pd.to_numeric(self.neutral_values.get(col), errors='coerce')
        self.calculation_df[col] = pd.to_numeric(self.calculation_df[col], errors='coerce')

    def _convert_boolean_column(self, col: str) -> None:
        """Converte coluna booleana e valores de refer칡ncia."""
        self.good_values[col] = str(self.good_values.get(col, '')).strip().upper() == 'TRUE'
        self.neutral_values[col] = str(self.neutral_values.get(col, '')).strip().upper() == 'TRUE'
        self.calculation_df[col] = (
            self.calculation_df[col]
            .astype(str)
            .str.strip()
            .str.upper()
            .map({'TRUE': True, 'FALSE': False})
            .fillna(False)
            .astype(bool)
        )

    def calculate_scores(self) -> None:
        """Calcula pontua칞칫es para todos os crit칠rios."""
        self._validate_calculation_preconditions()

        print("Colunas dispon칤veis para c치lculo:", list(self.calculation_df.columns))
        print("Pesos aplic치veis:", self.weights.to_dict())
        print("Tipos de Dados:", self.data_types.to_dict())

        for col in self.weights.index:
            self._process_column_for_scoring(col)

        self._calculate_total_score()

    def _validate_calculation_preconditions(self) -> None:
        """Verifica se os dados est칚o prontos para c치lculo."""
        if self.calculation_df is None or self.data_types is None or self.weights is None:
            raise ValueError("Dados n칚o carregados corretamente. Execute load_and_prepare_data() primeiro")

    def _process_column_for_scoring(self, col: str) -> None:
        """Processa uma coluna individual para c치lculo de pontua칞칚o."""
        print(f"\nProcessando coluna: {col}")

        score_col_name = f"{col}_score"
        dtype = self.data_types.get(col)

        try:
            if dtype in ['number', 'boolean']:
                self._calculate_numeric_score(col, score_col_name)
            elif dtype == 'string':
                self._calculate_string_score(col, score_col_name)

            self._show_partial_results(col, score_col_name)
        except Exception as e:
            print(f"Erro ao processar coluna '{col}': {str(e)}")
            self.calculation_df[score_col_name] = 0

    def _calculate_numeric_score(self, col: str, score_col_name: str) -> None:
        """Calcula pontua칞칚o para coluna num칠rica ou booleana."""
        good_value = self.good_values.get(col)
        neutral_value = self.neutral_values.get(col)
        weight = self.weights.get(col, 0)
        proportionality = self.proportionality.get(col)

        if pd.isna(good_value) or pd.isna(neutral_value) or not proportionality:
            print(f"Aviso: Valores de refer칡ncia ausentes para '{col}'. Pontua칞칚o zerada.")
            self.calculation_df[score_col_name] = 0
            return

        print(f"  Tipo: {self.data_types[col]}, Proporcionalidade: {proportionality}")
        print(f"  Valores BOM: {good_value}, NEUTRO: {neutral_value}, Peso: {weight}")

        self.calculation_df[score_col_name] = self.calculation_df.apply(
            lambda row: self._calculate_numeric_score_value(
                row[col], good_value, neutral_value, weight, proportionality
            ) if col in row else 0,
            axis=1
        )

    def _calculate_string_score(self, col: str, score_col_name: str) -> None:
        """Calcula pontua칞칚o para coluna de string."""
        if col not in self.calculation_df.columns:
            print(f"Erro: Coluna string '{col}' n칚o encontrada")
            self.calculation_df[score_col_name] = 0
            return

        pts_col = self.string_columns_map.get(col)
        print(f"  Tipo: string. Coluna de pontos esperada: {pts_col}. Peso: {self.weights.get(col, 0)}")

        self.calculation_df[score_col_name] = self.calculation_df.apply(
            lambda row: self._calculate_string_score_value(row[col], col),
            axis=1
        )

    def _show_partial_results(self, col: str, score_col_name: str) -> None:
        """Mostra resultados parciais para uma coluna."""
        if score_col_name in self.calculation_df:
            print(f"  Resultado parcial (Top 5):\n{self.calculation_df[[col, score_col_name]].head().to_string(index=False)}")
        else:
            print(f"  Aviso: Coluna de pontua칞칚o '{score_col_name}' n칚o foi criada.")

    def _calculate_total_score(self) -> None:
        """Calcula a pontua칞칚o total somando todas as pontua칞칫es individuais."""
        score_columns = [c for c in self.calculation_df.columns if c.endswith('_score')]

        if not score_columns:
            print("\nNenhuma coluna de pontua칞칚o gerada. Pontua칞칚o total n칚o calculada.")
            self.calculation_df['Total_Score'] = 0
        else:
            print(f"\nColunas usadas para Pontua칞칚o Total: {score_columns}")
            self.calculation_df['Total_Score'] = self.calculation_df[score_columns].sum(axis=1)

        print("\nResumo final das pontua칞칫es:")
        cols_to_show = [self.model_column] + score_columns + ['Total_Score']
        valid_cols = [c for c in cols_to_show if c in self.calculation_df.columns]
        print(self.calculation_df[valid_cols].head().to_string(index=False))

    def _calculate_numeric_score_value(
        self,
        value: Union[float, bool],
        good_value: Union[float, bool],
        neutral_value: Union[float, bool],
        weight: float,
        proportionality_type: str
    ) -> float:
        """Calcula a pontua칞칚o para um valor num칠rico ou booleano individual."""
        if isinstance(value, (bool, np.bool_)):
            return self._calculate_boolean_score(value, good_value, neutral_value, weight)

        if pd.isna(value) or pd.isna(good_value) or pd.isna(neutral_value) or pd.isna(weight):
            return 0

        if good_value == neutral_value:
            return 0

        if proportionality_type == 'proportional':
            return self._calculate_proportional_score(value, good_value, neutral_value, weight)
        elif proportionality_type == 'i_proportional':
            return self._calculate_inverse_proportional_score(value, good_value, neutral_value, weight)

        return 0

    def _calculate_boolean_score(
        self,
        value: bool,
        good_value: bool,
        neutral_value: bool,
        weight: float
    ) -> float:
        """Calcula pontua칞칚o para valores booleanos."""
        if good_value == neutral_value:
            return weight if value == good_value else 0
        else:
            if value == good_value:
                return weight
            elif value == neutral_value:
                return 0
            else:
                return -weight

    def _calculate_proportional_score(
        self,
        value: float,
        good_value: float,
        neutral_value: float,
        weight: float
    ) -> float:
        """Calcula pontua칞칚o proporcional."""
        if good_value > neutral_value:  # Caso crescente
            if value >= good_value:
                return weight
            elif value >= neutral_value:
                return ((value - neutral_value) / (good_value - neutral_value)) * weight
            else:
                denominator = neutral_value if neutral_value != 0 else 1
                return -((abs(neutral_value - value)) / abs(denominator)) * weight
        elif good_value < neutral_value:  # Caso decrescente
            if value <= good_value:
                return weight
            elif value <= neutral_value:
                return ((value - good_value) / (neutral_value - good_value)) * weight
            else:
                denominator = neutral_value if neutral_value != 0 else 1
                return -((abs(value - neutral_value)) / abs(denominator)) * weight
        else:
            return 0

    def _calculate_inverse_proportional_score(
        self,
        value: float,
        good_value: float,
        neutral_value: float,
        weight: float
    ) -> float:
        """Calcula pontua칞칚o inversamente proporcional."""
        if good_value < neutral_value:  # Caso decrescente normal
            if value <= good_value:
                return weight
            elif value <= neutral_value:
                return ((neutral_value - value) / (neutral_value - good_value)) * weight
            else:
                denominator = neutral_value if neutral_value != 0 else 1
                return -((abs(value - neutral_value)) / abs(denominator)) * weight
        elif good_value > neutral_value:  # Caso crescente incomum
            if value >= good_value:
                return weight
            elif value >= neutral_value:
                return ((good_value - value) / (good_value - neutral_value)) * weight
            else:
                denominator = neutral_value if neutral_value != 0 else 1
                return -((abs(neutral_value - value)) / abs(denominator)) * weight
        else:
            return 0

    def _calculate_string_score_value(self, value: str, column: str) -> float:
        """Calcula pontua칞칚o para um valor de string individual."""
        points_column = self.string_columns_map.get(column)

        if not points_column or points_column not in self.df.columns:
            return 0

        try:
            mapping_data = self.df.iloc[5:].copy()
            mapping_data[points_column] = pd.to_numeric(mapping_data[points_column], errors='coerce')
            mapping_data = mapping_data.dropna(subset=[column, points_column])

            mapping = pd.Series(
                mapping_data[points_column].values,
                index=mapping_data[column]
            ).to_dict()

            if not mapping:
                return 0

            clean_value = str(value).strip()
            base_score = mapping.get(clean_value, 0)
            column_weight = float(self.weights.get(column, 0))

            return base_score * column_weight
        except Exception as e:
            print(f"Erro ao calcular pontua칞칚o para '{column}': {str(e)}")
            return 0

    def generate_visualizations(self, pdf_export_path: Optional[str] = None) -> None:
        """Gera visualiza칞칫es e exporta para PDF se solicitado."""
        print("\n--- Gerando Visualiza칞칫es ---")

        if self.calculation_df is None:
            print("Dados n칚o dispon칤veis para gera칞칚o de gr치ficos")
            return

        try:
            # Lista para armazenar os caminhos das imagens tempor치rias
            temp_images = []

            # 1. Gr치fico de Pontua칞칚o Total
            total_score_fig = self._create_total_score_chart()
            if total_score_fig:
                self._display_chart(total_score_fig)
                if pdf_export_path:
                    temp_images.append(self._save_figure_as_png(total_score_fig, "total_score"))

            # 2. Gr치ficos Num칠ricos
            numeric_figs = self._create_numeric_charts()
            for i, fig in enumerate(numeric_figs):
                self._display_chart(fig)
                if pdf_export_path:
                    temp_images.append(self._save_figure_as_png(fig, f"numeric_{i}"))

            # 3. Gr치ficos de String
            string_figs = self._create_string_charts()
            for i, fig in enumerate(string_figs):
                self._display_chart(fig)
                if pdf_export_path:
                    temp_images.append(self._save_figure_as_png(fig, f"string_{i}"))

            # 4. Gr치ficos Booleanos
            boolean_figs = self._create_boolean_charts()
            for i, fig in enumerate(boolean_figs):
                self._display_chart(fig)
                if pdf_export_path:
                    temp_images.append(self._save_figure_as_png(fig, f"boolean_{i}"))

            # Combinar em PDF se necess치rio
            if pdf_export_path and temp_images:
                self._create_pdf_from_images(temp_images, pdf_export_path)
                print(f"Relat칩rio PDF gerado em: {pdf_export_path}")

            # Limpeza dos arquivos tempor치rios
            for img_path in temp_images:
                if os.path.exists(img_path):
                    os.remove(img_path)

        except Exception as e:
            print(f"Erro ao gerar visualiza칞칫es: {str(e)}")

    def _display_chart(self, fig: go.Figure) -> None:
        """Exibe um gr치fico interativo."""
        fig.show()

    def _get_colab_figure_size(self) -> Dict[str, int]:
        """Retorna as dimens칫es padr칚o do Google Colab"""
        return {
            'width': 800,   # Largura padr칚o do Colab
            'height': 500,  # Altura padr칚o do Colab
            'scale': 2      # Fator de escala para qualidade
        }

    def _create_total_score_chart(self) -> Optional[go.Figure]:
        """Cria gr치fico de pontua칞칚o total com tamanho do Colab"""
        if 'Total_Score' not in self.calculation_df.columns:
            return None

        # Cria칞칚o do gr치fico (c칩digo existente)
        fig = px.bar(
            self.calculation_df,
            x='Total_Score',
            y=self.model_column,
            orientation='h',
            # ... outros par칙metros ...
        )

        # Aplica o tamanho do Colab
        size = self._get_colab_figure_size()
        fig.update_layout(
            width=size['width'],
            height=size['height'],
            autosize=False
        )

        return fig

    def _save_figure_as_png(self, fig: go.Figure, name: str) -> str:
        """Salva figura como PNG com tamanho do Colab"""
        try:
            with tempfile.NamedTemporaryFile(suffix=".png", prefix=name, delete=False) as tmpfile:
                img_path = tmpfile.name
                size = self._get_colab_figure_size()

                fig.write_image(
                    img_path,
                    width=size['width'],
                    height=size['height'],
                    scale=size['scale'],
                    engine="kaleido"
                )
                return img_path
        except Exception as e:
            print(f"Erro ao salvar figura {name}: {str(e)}")
            return ""

    def _create_pdf_from_images(self, image_paths: List[str], output_pdf: str) -> None:
        """Combina imagens em PDF mantendo propor칞칚o do Colab"""
        try:
            pdf = FPDF(orientation='L')
            colab_ratio = 800/500  # Propor칞칚o largura/altura do Colab

            for img_path in image_paths:
                if not os.path.exists(img_path):
                    continue

                pdf.add_page()

                # Calcula dimens칫es mantendo propor칞칚o
                pdf_w = pdf.w - 40
                pdf_h = pdf.w / colab_ratio

                if pdf_h > (pdf.h - 40):
                    pdf_h = pdf.h - 40
                    pdf_w = pdf_h * colab_ratio

                # Centraliza na p치gina
                x = (pdf.w - pdf_w) / 2
                y = (pdf.h - pdf_h) / 2

                pdf.image(img_path, x=x, y=y, w=pdf_w, h=pdf_h)

            pdf.output(output_pdf)

        except Exception as e:
            print(f"Erro ao criar PDF: {str(e)}")
            raise

    def _add_reference_annotation(
        self,
        fig: go.Figure,
        x_value: float,
        y_position: int,
        text: str
    ) -> None:
        """Adiciona anota칞칚o de refer칡ncia ao gr치fico."""
        fig.add_annotation(
            x=x_value,
            y=y_position - 1,
            xref="x",
            yref="y",
            text=text,
            showarrow=True,
            font=dict(size=12, color="#000000"),
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            arrowcolor="#636363",
            ax=0,
            ay=-40,
            bordercolor="#c7c7c7",
            borderwidth=1,
            borderpad=4,
            bgcolor='white',
        )

    def _add_reference_line(
        self,
        fig: go.Figure,
        value: float,
        y_position: int,
        text: str,
        is_inverse: bool = False
    ) -> None:
        """Adiciona linha de refer칡ncia ao gr치fico com anota칞칚o."""
        fig.add_vline(
            x=value,
            line_dash="dash",
            line_color="gray",
            line_width=1,
        )

        fig.add_annotation(
            x=value,
            y=y_position - 1,
            text=text,
            showarrow=True,
            font=dict(size=12, color="#000000"),
            arrowhead=2,
            arrowsize=1,
            arrowwidth=2,
            arrowcolor="#636363",
            ax=0,
            ay=-40,
            bordercolor="#c7c7c7",
            borderwidth=1,
            borderpad=4,
            bgcolor='white',
        )

    def _classify_numeric_value(
        self,
        value: float,
        good_value: float,
        neutral_value: float,
        is_inverse: bool
    ) -> str:
        """Classifica um valor num칠rico como Positivo, Negativo ou Neutro."""
        if pd.isna(neutral_value):
            return 'Neutra'

        value = float(value)
        neutral_value = float(neutral_value)

        if is_inverse:
            return 'Positiva' if value <= neutral_value else 'Negativa'
        else:
            return 'Positiva' if value >= neutral_value else 'Negativa'

    def generate_visualizations(self, pdf_export_path: Optional[str] = None) -> None:
        """Gera todas as visualiza칞칫es de an치lise e opcionalmente exporta para PDF."""
        print("\n--- Gerando Visualiza칞칫es ---")

        if self.calculation_df is None:
            print("Dados n칚o dispon칤veis para gera칞칚o de gr치ficos")
            return

        try:
            temp_images = []

            total_score_fig = self._generate_total_score_chart()
            if pdf_export_path and total_score_fig:
                temp_images.append(self._save_figure_as_png(total_score_fig, "total_score"))

            numeric_figs = self._generate_numeric_charts()
            if pdf_export_path and numeric_figs:
                temp_images.extend(self._save_figures(numeric_figs, "numeric"))

            string_figs = self._generate_string_charts()
            if pdf_export_path and string_figs:
                temp_images.extend(self._save_figures(string_figs, "string"))

            boolean_figs = self._generate_boolean_charts()
            if pdf_export_path and boolean_figs:
                temp_images.extend(self._save_figures(boolean_figs, "boolean"))

            if pdf_export_path and temp_images:
                self._combine_pngs_to_pdf(temp_images, pdf_export_path)
                for img_path in temp_images:
                    if os.path.exists(img_path):
                        os.remove(img_path)

            print("--- Visualiza칞칫es Geradas ---")
        except Exception as e:
            print(f"Erro ao gerar visualiza칞칫es: {str(e)}")
            if 'temp_images' in locals():
                for img_path in temp_images:
                    if os.path.exists(img_path):
                        os.remove(img_path)

    def _save_figures(self, figures: List[go.Figure], prefix: str) -> List[str]:
        """Salva m칰ltiplas figuras como PNGs tempor치rios."""
        temp_images = []
        for i, fig in enumerate(figures):
            img_path = self._save_figure_as_png(fig, f"{prefix}_{i}")
            if img_path:
                temp_images.append(img_path)
        return temp_images

    def _save_figure_as_png(self, fig: go.Figure, name: str) -> str:
        """Salva uma figura Plotly como PNG tempor치rio."""
        try:
            with tempfile.NamedTemporaryFile(suffix=".png", prefix=name, delete=False) as tmpfile:
                img_path = tmpfile.name
                fig.write_image(
                    img_path,
                    width=1200,
                    height=600,
                    scale=2,
                    engine="kaleido"
                )
            return img_path
        except Exception as e:
            print(f"Erro ao salvar figura {name}: {str(e)}")
            return ""

    def _combine_pngs_to_pdf(self, image_paths: List[str], output_pdf: str) -> None:
        """Combina v치rias imagens PNG em um 칰nico PDF."""
        try:
            pdf = FPDF(orientation='L')

            for img_path in image_paths:
                if not os.path.exists(img_path):
                    continue

                pdf.add_page()

                with Image.open(img_path) as img:
                    img_w, img_h = img.size
                    aspect = img_w / img_h

                    pdf_w = pdf.w - 40
                    pdf_h = pdf.h - 40

                    if (pdf_w / pdf_h) < aspect:
                        width = pdf_w
                        height = width / aspect
                    else:
                        height = pdf_h
                        width = height * aspect

                    x = (pdf.w - width) / 2
                    y = (pdf.h - height) / 2

                    pdf.image(img_path, x=x, y=y, w=width, h=height)

            pdf.output(output_pdf)
            print(f"PDF gerado com sucesso: {output_pdf}")
        except Exception as e:
            print(f"Erro ao gerar PDF: {str(e)}")
            raise

    def _generate_total_score_chart(self) -> Optional[go.Figure]:
        """Gera gr치fico de barras para pontua칞칚o total."""
        if 'Total_Score' not in self.calculation_df.columns:
            print("Pontua칞칚o total n칚o calculada. Pulando gr치fico.")
            return None

        max_theoretical_score = self.weights.sum() if self.weights is not None else 1.0
        print(f"  Pontua칞칚o m치xima te칩rica: {max_theoretical_score:.2f}")

        analysis_df = self.calculation_df[
            ~self.calculation_df[self.model_column].isin(['BOM', 'NEUTRO'])
        ].copy()

        analysis_df.dropna(subset=['Total_Score'], inplace=True)

        if analysis_df.empty:
            print("Nenhum dado v치lido para plotar")
            return None

        analysis_df = analysis_df.sort_values('Total_Score', ascending=False)
        analysis_df['Color_Category'] = analysis_df['Total_Score'].apply(
            lambda x: 'Positiva' if x >= 0 else 'Negativa'
        )

        fig = px.bar(
            analysis_df,
            x='Total_Score',
            y=self.model_column,
            orientation='h',
            color='Color_Category',
            color_discrete_map={'Positiva': 'lightgreen', 'Negativa': 'palevioletred'},
            hover_name=self.model_column,
            hover_data={'Total_Score': ':.2f'},
            title='Pontua칞칚o Total dos Modelos'
        )

        fig.update_layout(
            yaxis={'categoryorder': 'total ascending'},
            plot_bgcolor='white',
            paper_bgcolor='white',
            legend_title_text='Resultado',
            xaxis_title='Pontua칞칚o Total',
            yaxis_title='Modelo'
        )

        fig.add_vline(x=0, line_dash='dash', line_color='black', line_width=1)
        fig.add_vline(x=max_theoretical_score, line_dash='dash', line_color='black', line_width=1)

        self._add_reference_annotation(
            fig, max_theoretical_score, len(analysis_df),
            "Desej치vel (BOM)<br>(NEUTRO)"
        )
        self._add_reference_annotation(
            fig, 0, len(analysis_df),
            "M칤nimo Aceit치vel<br>(NEUTRO)"
        )

        if not analysis_df.empty:
            best_product = analysis_df.iloc[0]
            fig.add_annotation(
                x=best_product['Total_Score'],
                y=best_product[self.model_column],
                text="游끥 Melhor Produto",
                showarrow=False,
                font=dict(color="#00008B", size=12),
                bgcolor="white",
                bordercolor="black",
                borderwidth=1,
                borderpad=4,
                xshift=-15,
                yshift=25
            )

        fig.show()
        return fig

    def _generate_numeric_charts(self) -> List[go.Figure]:
        """Gera gr치ficos para colunas num칠ricas."""
        figures = []

        if self.data_types is None:
            return figures

        numeric_cols = [
            col for col in self.data_types.index
            if self.data_types[col] == 'number' and col in self.calculation_df.columns
        ]

        for col in numeric_cols:
            print(f"Gerando gr치fico para coluna num칠rica: '{col}'")

            df = self.calculation_df[
                ~self.calculation_df[self.model_column].isin(['BOM', 'NEUTRO'])
            ][[self.model_column, col]].copy()

            df.dropna(subset=[col], inplace=True)

            if df.empty:
                continue

            good_value = self.good_values.get(col)
            neutral_value = self.neutral_values.get(col)
            is_inverse = self.proportionality.get(col, '').lower() == 'i_proportional'

            df['Color_Category'] = df.apply(
                lambda x: self._classify_numeric_value(
                    x[col], good_value, neutral_value, is_inverse
                ),
                axis=1
            )

            df = df.sort_values(col, ascending=not is_inverse)

            fig = px.bar(
                df,
                x=col,
                y=self.model_column,
                orientation='h',
                color='Color_Category',
                color_discrete_map={
                    'Positiva': 'lightgreen',
                    'Negativa': 'palevioletred',
                    'Neutra': 'lightgray'
                },
                hover_name=self.model_column,
                hover_data={col: ':.2f'},
                title=f"{col} {'(Inversamente Proporcional)' if is_inverse else ''}"
            )

            fig.update_layout(
                plot_bgcolor='white',
                paper_bgcolor='white',
                legend_title_text='Resultado',
                xaxis_title=col,
                yaxis_title='Modelo'
            )

            if pd.notna(neutral_value):
                self._add_reference_line(
                    fig, float(neutral_value), len(df),
                    "M칤nimo Aceit치vel<br>(NEUTRO)",
                    is_inverse
                )
            if pd.notna(good_value):
                self._add_reference_line(
                    fig, float(good_value), len(df),
                    "Desej치vel (BOM)",
                    is_inverse
                )

            fig.show()
            figures.append(fig)

        return figures

    def _generate_string_charts(self) -> List[go.Figure]:
        """Gera gr치ficos para colunas de string."""
        figures = []

        if self.data_types is None:
            return figures

        string_cols = [
            col for col in self.data_types.index
            if self.data_types.get(col) == 'string' and col in self.calculation_df.columns
        ]

        for col in string_cols:
            print(f"Gerando gr치fico para string: '{col}'")
            pts_col = self.string_columns_map.get(col)

            if not pts_col or pts_col not in self.df.columns:
                continue

            try:
                mapping_data = self.df.iloc[5:].copy()
                mapping_data[pts_col] = pd.to_numeric(mapping_data[pts_col], errors='coerce')
                mapping_data = mapping_data.dropna(subset=[col, pts_col])

                string_mapping = pd.Series(
                    mapping_data[pts_col].values,
                    index=mapping_data[col]
                ).to_dict()

                if not string_mapping:
                    continue

                df = self.calculation_df[
                    ~self.calculation_df[self.model_column].isin(['BOM', 'NEUTRO'])
                ][[self.model_column, col]].copy()

                df['BaseValue'] = df[col].map(string_mapping).fillna(0)
                df = df.sort_values('BaseValue', ascending=False)
                df['Position'] = range(len(df))

                fig = px.scatter(
                    df,
                    x='Position',
                    y='BaseValue',
                    color=col,
                    hover_name=self.model_column,
                    title=f'Compara칞칚o de {col.capitalize()} entre Modelos (Pontua칞칚o Base)',
                    size_max=15,
                    hover_data={'Position': False, 'BaseValue': True, col: True},
                    labels={
                        'BaseValue': 'Pontua칞칚o Base (do CSV)',
                        col: col.capitalize()
                    }
                )

                good_str = self.good_values.get(col)
                neutral_str = self.neutral_values.get(col)
                good_base = string_mapping.get(good_str)
                neutral_base = string_mapping.get(neutral_str)

                if good_base is not None:
                    fig.add_hline(
                        y=good_base,
                        line_dash='dash',
                        line_color='green',
                        line_width=2,
                        annotation_text=f"BOM: '{good_str}' ({good_base:.2f})",
                        annotation_position="top right",
                        annotation_font=dict(color='green', size=12)
                    )

                if neutral_base is not None:
                    fig.add_hline(
                        y=neutral_base,
                        line_dash='dash',
                        line_color='orange',
                        line_width=2,
                        annotation_text=f"NEUTRO: '{neutral_str}' ({neutral_base:.2f})",
                        annotation_position="bottom right",
                        annotation_font=dict(color='orange', size=12)
                    )

                fig.update_layout(
                    xaxis=dict(
                        title='Modelos Ordenados por Pontua칞칚o Base',
                        showticklabels=False,
                        showgrid=False,
                        zeroline=False
                    ),
                    yaxis=dict(
                        title='Pontua칞칚o Base (Definida no CSV)',
                        showgrid=True,
                        gridcolor='lightgray',
                        zeroline=True,
                        zerolinecolor='lightgray'
                    ),
                    plot_bgcolor='white',
                    paper_bgcolor='white',
                    margin=dict(l=40, r=40, t=60, b=40),
                    legend=dict(
                        title_text=col.capitalize(),
                        orientation='h',
                        yanchor='bottom',
                        y=-0.3,
                        xanchor='center',
                        x=0.5
                    ),
                    hoverlabel=dict(
                        bgcolor='white',
                        font_size=12,
                        font_family='Arial'
                    )
                )

                fig.update_traces(
                    marker=dict(
                        size=12,
                        line=dict(width=1, color='DarkSlateGrey'),
                    )
                )

                fig.show()
                figures.append(fig)
            except Exception as e:
                print(f"Erro ao gerar gr치fico para '{col}': {str(e)}")

        return figures

    def _generate_boolean_charts(self) -> List[go.Figure]:
        """Gera gr치ficos para colunas booleanas."""
        figures = []

        if self.data_types is None:
            return figures

        bool_cols = [
            col for col in self.data_types.index
            if self.data_types[col] == 'boolean' and col in self.calculation_df.columns
        ]

        for col in bool_cols:
            df = self.calculation_df[
                ~self.calculation_df[self.model_column].isin(['BOM', 'NEUTRO'])
            ][[self.model_column, col]].copy()

            df['Value'] = df[col].map({True: 1, False: 0})
            df['Size'] = 20

            fig = px.scatter(
                df,
                x=self.model_column,
                y='Value',
                color=col,
                color_discrete_map={True: '#4CAF50', False: '#F44336'},
                size='Size',
                title=f'{col} por Modelo',
                hover_name=self.model_column,
                hover_data={'Value': False, 'Size': False},
                category_orders={col: [True, False]}
            )

            fig.for_each_trace(
                lambda t: t.update(name='SIM' if t.name == 'True' else 'N츾O')
            )

            fig.update_yaxes(
                range=[-0.5, 1.5],
                tickvals=[0, 1],
                ticktext=['N칚o', 'Sim'],
                showgrid=False
            )

            fig.add_hline(y=1, line_dash='dash', line_color='green')
            fig.add_hline(y=0, line_dash='dash', line_color='orange')

            fig.update_layout(
                xaxis_title=None,
                yaxis_title=None,
                plot_bgcolor='white',
                paper_bgcolor='white'
            )

            fig.show()
            figures.append(fig)

        return figures

    def recommend_products(self) -> None:
        """Recomenda produtos com base na an치lise."""
        if 'Total_Score' not in self.calculation_df.columns:
            print("Pontua칞칚o total n칚o calculada. N칚o 칠 poss칤vel recomendar.")
            return

        products = self.calculation_df[~self.calculation_df[self.model_column].isin(['BOM', 'NEUTRO'])]

        if products.empty:
            print("Nenhum produto encontrado para recomenda칞칚o.")
            return

        max_score = products['Total_Score'].max()

        if pd.isna(max_score):
            print("N칚o foi poss칤vel determinar a pontua칞칚o m치xima.")
            return

        recommended = products[products['Total_Score'] == max_score]

        print("\n--- Produtos Recomendados ---")

        if recommended.empty:
            print("Nenhum produto atingiu a pontua칞칚o m치xima.")
            return

        for _, product in recommended.iterrows():
            print(f"\nProduto: {product[self.model_column]} (Pontua칞칚o: {product['Total_Score']:.2f})")
            print("Crit칠rios destacados:")

            # Ordena por peso dos crit칠rios
            sorted_scores = sorted(
                [(col, product[f"{col}_score"])
                 for col in self.weights.index
                 if f"{col}_score" in product],
                key=lambda x: self.weights[x[0]],
                reverse=True
            )

            for col, score in sorted_scores[:5]:  # Mostra apenas os 5 principais
                print(f"- {col}: {score:.2f} (Peso: {self.weights[col]:.2f})")


def analyze_electronics_data(file_path: str = 'base_eletronicos_2.csv', pdf_output: Optional[str] = None) -> None:
    """Executa an치lise completa para dados de eletr칪nicos."""
    try:
        print("--- An치lise para base de eletr칪nicos ---")
        analyzer = DataAnalyzer(file_path)
        analyzer.load_and_prepare_data()
        analyzer.calculate_scores()
        analyzer.generate_visualizations(pdf_export_path=pdf_output)
        analyzer.recommend_products()
    except FileNotFoundError:
        print("\nErro: Arquivo n칚o encontrado.")
    except Exception as e:
        print(f"\nOcorreu um erro inesperado: {str(e)}")


if __name__ == "__main__":
    analyze_electronics_data(pdf_output='relatorio_analise.pdf')
```
